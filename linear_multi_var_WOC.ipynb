```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv("linear_train.csv")
df
```

```python
df=df.drop("Unnamed: 0",axis=1)
```

```python
x_train=np.array(df.loc[0:,"A":"T"])
y_train=np.array(df.loc[0:,'label'])
x_train.shape
```

```python
mu=np.mean(x_train,axis=0)
sigma=np.std(x_train,axis=0)
x_train=(x_train-mu)/sigma
x_train
mu1=np.mean(y_train)
sigma1=np.std(y_train)
y_train=(y_train-mu1)/sigma1
y_train
```

```python
def cost_fun(x,y,w,b):
    cost=0
    m=x.shape[0]
    y_pred=np.dot(x,w)+b
    cost+=(1/(2*m))*np.sum((y_pred-y)**2)
    return cost
```

```python
def linear_reg(x,y,alpha,iters):
    w=np.ones(x.shape[1],)*0.5
    
    b=1
    
    m=x.shape[0]
    for i in range(iters):
        cost1=cost_fun(x,y,w,b)
        y_pred=np.dot(x,w)+b
        dj_dw=(1/m)*np.dot(x.T,y_pred-y)
        w=w-alpha*dj_dw
        dj_db=(1/m)*np.sum(y_pred-y)
        b=b-alpha*dj_db
        if(i%10==0):
            print(f'cost:{cost1},i:{i}')
    return w,b,cost1
        
```

```python
w2,b1,cost2=linear_reg(x_train,y_train,0.005,10000)
```

```python
w2
```

```python
y_pred = np.dot(x_train, w2)+b1
error = (1/x_train.shape[0])*np.sum(y_pred - y_train)
```

```python
accuracy=(1-error)*100
print(accuracy)
```
