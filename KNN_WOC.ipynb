```python
distance=np.zeros((x_test.shape[0],x_train.shape[0]))
for i in range(x_train.shape[0]):
    for j in range(x_test.shape[0]):
        distance[j][i]=np.sqrt(np.sum((x_train[i]-x_test[j])**2))
```

```python
distance1=distance.argsort()
distance1
```
